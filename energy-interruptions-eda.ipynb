```python
def string_search(s, search):
    return search in str(s)

df_temp = pd.DataFrame(data = df_concat2['Area Affected'].values)
#df_temp = df_concat2['Area Affected']

mask2 = df_temp.applymap(lambda x: string_search(x, ':'))
mask2
```

```python
def split_string(series):
    return re.findall("([A-Za-z]+)\:",series)
```

```python
df_temp = df_temp.loc[mask2.any(axis=1)].applymap(lambda x: split_string(x))
df_temp
```
