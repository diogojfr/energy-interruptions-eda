### Correcting columns 

```python
# replacing white space with underscore
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
# dataset dtypes
df.info()
```

```python
df.isnull().sum()
```

```python
# dropping the 'alert criteria' and 'Nerc Region' columns
df = df.drop(columns=['NERC_Region',
                      'Alert_Criteria'])
df.head()
```

```python
# There is a double January month
df['Month'].unique()
```

```python
# Replacing the wrong january months
df.loc[df['Month'] == 'January ','Month'] = 'January'
df['Month'].unique()
```

```python
# Concatenating the data and time events in one column

df['Date_Time_Event_Began'] = df['Date_Event_Began'] +' '+ df['Time_Event_Began']
df['Date_Time_Restoration'] = df['Date_of_Restoration'] +' '+ df['Time_of_Restoration']

# Dropping the old columns
df = df.drop(columns=['Date_Event_Began','Time_Event_Began','Date_of_Restoration','Time_of_Restoration'])

df.head()
```

```python
# Setting the date/time events colmuns as datetime format
df['Date_Time_Event_Began'] = pd.to_datetime(df['Date_Time_Event_Began'], errors='coerce')
df['Date_Time_Restoration'] = pd.to_datetime(df['Date_Time_Restoration'], errors='coerce')

df.info()
```

```python
# It will be removed the counties names from the columns
# Only the state affected will be stored
df['Area_Affected'].unique()
```

```python
# 1. finding all the lines with contains ':', i.e., lines with county names
# 2. Spliting the strings at the ':' delimiter
# 3. Getting the firt position of the list, i.e., the state name

state_names_extracted = df.loc[df['Area_Affected'].str.contains(':'),'Area_Affected'] \
    .str.split(':')\
    .str.get(0)

# substituing the state names in the dataframe
df.loc[df['Area_Affected'].str.contains(':'),'Area_Affected'] = state_names_extracted
df['Area_Affected'].unique()
```

```python
# Correcting the state names
df.loc[df['Area_Affected']=='District of Columbia'] = 'Washington'
df.loc[df['Area_Affected']=='Northern and Central California;'] = 'California'
df.loc[df['Area_Affected']=='Western NY'] = 'New York'
df.loc[df['Area_Affected']=='Central Oklahoma'] = 'Oklahoma'
df.loc[df['Area_Affected']=='Texas panhandle, SE NM'] = 'Texas, New Mexico'

df['Area_Affected'].unique()
```
