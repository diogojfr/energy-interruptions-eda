### Importing the concatenated dataframe

### Basic information about the dataset

```python
#dataset shape
print(f'The dataset has {df.shape[0]} rows and {df.shape[1]} columns.')
```

```python
# dataset columns
df.columns
```

### Removing the duplicated data

```python
# verifying the duplicates
df.duplicated().sum()
```

```python
# removing the duplicates
mask = df.duplicated()
df = df[~mask]
df.duplicated().sum()
```

### Correcting columns names

```python
# replacing white space with underscore
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
# dataset dtypes
df.info()
```

```python
df.isnull().sum()
```

### Dropping the 'alert criteria' and 'Nerc Region' columns

```python

df = df.drop(columns=['NERC_Region',
                      'Alert_Criteria'])
df.head()
```

### Correcting the Month column

```python
# There is a double January month
df['Month'].unique()
```

```python
# Replacing the wrong january months
df.loc[df['Month'] == 'January ','Month'] = 'January'
df['Month'].unique()
```

### Concatenating date and time events in one column


```python

df['Date_Time_Event_Began'] = df['Date_Event_Began'] +' '+ df['Time_Event_Began']
df['Date_Time_Restoration'] = df['Date_of_Restoration'] +' '+ df['Time_of_Restoration']

# Dropping the old columns
df = df.drop(columns=['Date_Event_Began','Time_Event_Began','Date_of_Restoration','Time_of_Restoration'])

df.head()
```

```python
# Setting the date/time events colmuns as datetime format
df['Date_Time_Event_Began'] = pd.to_datetime(df['Date_Time_Event_Began'], errors='coerce')
df['Date_Time_Restoration'] = pd.to_datetime(df['Date_Time_Restoration'], errors='coerce')

df.info()
```

### Correcting the 'Area_Affected' column

```python
# It will be removed the counties names from the columns
# Only the state affected will be stored
df['Area_Affected'].unique()
```

```python
# 1. finding all the lines with contains ':', i.e., lines with county names
# 2. Spliting the strings at the ':' delimiter
# 3. Getting the firt position of the list, i.e., the state name

state_names_extracted = df.loc[df['Area_Affected'].str.contains(':'),'Area_Affected'] \
    .str.split(':')\
    .str.get(0)

# substituing the state names in the dataframe
df.loc[df['Area_Affected'].str.contains(':'),'Area_Affected'] = state_names_extracted
df['Area_Affected'].unique()
```

```python
# Correcting the state names
df.loc[df['Area_Affected']=='District of Columbia','Area_Affected'] = 'Washington'
df.loc[df['Area_Affected']=='Northern and Central California;','Area_Affected'] = 'California'
df.loc[df['Area_Affected']=='Western NY','Area_Affected'] = 'New York'
df.loc[df['Area_Affected']=='Central Oklahoma','Area_Affected'] = 'Oklahoma'
df.loc[df['Area_Affected']=='Texas panhandle, SE NM','Area_Affected'] = 'Texas, New Mexico'

df['Area_Affected'].unique()
```

### Replacing 'unknown' by NaN values

```python
# Replacing 'unknown' by NaN values

df = df.replace(to_replace='Unknown', value=np.nan)
df.head()
```

### Correcting the 'Event type' column

```python
df['Event_Type'].value_counts()
```

```python
# correcting the 'Sever Weather' typo
df.loc[df['Event_Type'] == 'Sever Weather','Event_Type'] = 'Severe Weather'

# correcting  'Suspicious activity'
df.loc[df['Event_Type'] == 'Suspicious activity','Event_Type'] = 'Suspicious Activity'

# correcting the 'Transmisison Interruption' typo
df.loc[df['Event_Type'] == 'Severe Weather/Transmisison Interruption','Event_Type'] = 'Severe Weather/Transmission Interruption'

df['Event_Type'].value_counts()
```

### Adding the 'Event_Duration' column

```python
# calculating the difference between the initial and restoration dates
df['Event_Duration_(hours)'] = df['Date_Time_Restoration'] - df['Date_Time_Event_Began']

# tranforming the format deltatime to hours, rounding the result to only two decimals
df['Event_Duration_(hours)'] = [round(x.days*24 + x.seconds/(60*60),2) for x in df['Event_Duration_(hours)']]

df.head()
```

```python
# There is an negative duration
# probably the date of restoration was entered incorrectly
df.loc[df['Event_Duration_(hours)'] <= 0]
```

```python
# fixing the date of restoration modifying the year to 2019
df.loc[df['Date_Time_Restoration'] == '2018-08-18 23:00:00','Date_Time_Restoration'] = '2019-08-18 23:00:00'

# creating again the event duration column
df['Event_Duration_(hours)'] = df['Date_Time_Restoration'] - df['Date_Time_Event_Began']
df['Event_Duration_(hours)'] = [round(x.days*24 + x.seconds/(60*60),2) for x in df['Event_Duration_(hours)']]

# checking negatives durations
df.loc[df['Event_Duration_(hours)'] <= 0]
```

### Converting the 'demand loss' and 'number of customers affected' columns to float type

```python
df.info()
```

```python
df['Demand_Loss_(MW)'] = df['Demand_Loss_(MW)'].astype('float64')
df['Number_of_Customers_Affected'] = df['Number_of_Customers_Affected'].astype('float64')

df.info()
```

### Adding the 'year' column

```python
df['Year'] = [x.year for x in df['Date_Time_Event_Began']]
df.head()
```

```python
# checking the 'year' column
df['Year'].value_counts()
```

### Saving the final dataframe to a csv file

```python
df.to_csv('datasets/df.csv', index=False)
```
